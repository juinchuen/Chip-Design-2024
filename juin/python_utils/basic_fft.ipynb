```python
t = np.linspace(0,1,64)
pi = 3.1415

x = np.round((np.sin(2 * pi * t * 4) + np.cos(2 * pi * t * 16)) * 512)

plt.plot(t, x)
```

```python
y = np.fft.fft(x)

plt.plot(y.real)
plt.plot(y.imag)
```
